### 1.1. Скачать данные

```python
import pandas as pd
df = pd.read_csv('kc_house_data.csv', sep=',')
df.head()
```

### 1.2 Изучите стоимости недвижимости

```python
import matplotlib.pyplot as plt

plt.hist(df['price'])
plt.title('Распределение стоимости')
plt.xlabel('price')
plt.ylabel('Колличество');
```

##### Вывод: основная чать недвижимости имеет невысокую стоимость

### 1.3 Изучите распределение квадратуры жилой

```python
plt.hist(df['sqft_living'])
plt.title('Распределение жилой площади')
plt.xlabel('sqft_living')
plt.ylabel('Колличество');
```

##### Вывод: основная чать недвижимости имеет жилую площадь до 4000

### 1.4 Изучите распределение года постройки

```python
plt.hist(df['yr_built'])
plt.title('Распределение года постройки')
plt.xlabel('yr_built')
plt.ylabel('Колличество');
```

##### Вывод: наибольшая чать недвижимости построена в период 2010-2020 год, но так же много недвижимости постороенной в период  1950-2000 год

## Задача 2

### 2.1 Изучите распределение домов от наличия вида на набережную
- Постройте график
- Сделайте выводы

```python
data = df['waterfront'].value_counts()
names = data.index
values = data.values

data
```

```python
plt.pie(values, autopct='%.1f%%', labels=names);
```

##### Вывод: 0.8% домов имеет вид на набережную

### 2.2 Изучите распределение этажей домов

##### Вывод: Практичски 50% недвижимости одноэтажное, на втором месте 2-х этажные дома.

## Задача 3
- Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
- Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
import numpy as np
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] =0
corr_matrix
```

```python
import seaborn as sns
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```

```python
sns.jointplot(x=df['sqft_living'], y=df['price'], kind='reg');
```

```python
sns.jointplot(x=df['grade'], y=df['price'], kind='reg');
```

```python
plt.figure(figsize=(16, 8))
sns.boxplot(x=df['price'], y=df['waterfront'].astype('str'), whis=1.5)

plt.title('distribution of price by waterfront')
plt.xlabel('price')
plt.ylabel('waterfront');
```

```python
plt.figure(figsize=(16, 8))
sns.boxplot(x=df['price'], y=df['view'].astype('str'), whis=1.5)

plt.title('distribution of price by waterfront')
plt.xlabel('price')
plt.ylabel('view');
```

```python
plt.figure(figsize=(16, 8))
sns.boxplot(x=df['price'], y=df['grade'].astype('str'), whis=1.5)

plt.title('distribution of price by waterfront')
plt.xlabel('price')
plt.ylabel('grade');
```
